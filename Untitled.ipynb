```python
train_dog = []
train_dog_target = []
dog_images = os.listdir('train/dogs/')
for i in dog_images:
    try:
        image = mpimg.imread(f'train/dogs/{i}')
        image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        image = cv2.resize(image, (300, 300))
        image = image.reshape(300, 300, 1)
        train_dog.append(image)
        train_dog_target.append('DOG')
    except:
        print(i)
train_dog = np.array(train_dog)
train_dog_target = np.array(train_dog_target)
```

```python
train_cat = []
train_cat_target = []
cat_images = os.listdir('train/cats/')
for i in cat_images:
    try:
        image = mpimg.imread(f'train/cats/{i}')
        image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        image = cv2.resize(image, (300, 300))
        image = image.reshape(300, 300, 1)
        train_cat.append(image)
        train_cat_target.append('CAT')
    except:
        print(i)
train_cat = np.array(train_cat)
train_cat_target = np.array(train_cat_target)
```

```python
test_dog = []
test_dog_target = []
dog_images = os.listdir('test/dogs/')
for i in dog_images:
    try:
        image = mpimg.imread(f'test/dogs/{i}')
        image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        image = cv2.resize(image, (300, 300))
        image = image.reshape(300, 300, 1)
        test_dog.append(image)
        test_dog_target.append('DOG')
    except:
        print(i)
test_dog = np.array(test_dog)
test_dog_target = np.array(test_dog_target)
```

```python
test_cat = []
test_cat_target = []
cat_images = os.listdir('test/cats/')
for i in cat_images:
    try:
        image = mpimg.imread(f'test/cats/{i}')
        image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        image = cv2.resize(image, (300, 300))
        image = image.reshape(300, 300, 1)
        test_cat.append(image)
        test_cat_target.append('CAT')
    except:
        print(i)
test_cat = np.array(test_cat)
test_cat_target = np.array(test_cat_target)
```

```python
X_train = np.concatenate([train_dog, train_cat])
```

```python
train_dog.shape
```

```python
train_cat.shape
```

```python
X_train.shape
```

```python
y_train = np.concatenate([train_dog_target, train_cat_target])
```

```python
y_train.shape
```

```python
X_test = np.concatenate([test_dog, test_cat])
```

```python
y_test = np.concatenate([test_dog_target, test_cat_target])
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
from keras.utils import to_categorical
```

```python
y_train = pd.get_dummies(y_train).astype(int).values
y_test = pd.get_dummies(y_test).astype(int).values
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
from keras import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
```

```python
model = Sequential()

model.add(Conv2D(filters=64, kernel_size=(5, 5), padding='same', activation='relu', input_shape = (300, 300, 1)))
model.add(MaxPooling2D(pool_size=(4, 4)))

model.add(Conv2D(filters=32, kernel_size=(3, 3), padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())

model.add(Dense(32, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(2, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=10)
```
